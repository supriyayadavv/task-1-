#HISTOGRAM PLOT FOR AGE 

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#Box plot for Annual Salary across Departments

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Department', y='Annual Salary', data=data)
plt.title('Annual Salary across Departments')
plt.xlabel('Department')
plt.xticks(rotation=45)  # Rotate x-labels for better readability
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Department', y='Age', data=data)
plt.title('Age Distribution across Departments')
plt.xlabel('Department')
plt.ylabel('Age')
plt.show()
```

```python
pivot_table = data.groupby(['Gender', 'Department']).size().unstack().fillna(0)

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='g')
plt.title('Gender vs Department Co-occurrences')
plt.xlabel('Department')
plt.ylabel('Gender')
plt.show()
```
